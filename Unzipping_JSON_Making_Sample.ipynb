# Сохраняем Госкаталог

Само перекодирование Госкаталога из отдельных JSON в CSV происходит в коде Дмитрия на R. Здесь мы разархивируем Zip-хранилище JSON по частям в папку на рабочем компьютере, потом делаем выборку в 1%, потом делаем выборку в 1% из отдельного файла с изображениями, потом выгружаем отдельную колонку ("Технологии и материалы") полной таблицы. 

```python
from zipfile import ZipFile
import pandas as pd
import random

import csv
from io import TextIOWrapper
```

```python
# разархивация по частям

counter = 0

with ZipFile('data-4-structure-3.json.zip') as zf:
    for name in zf.namelist():
        counter += 1
        if counter <= 2000:
            continue
        if counter > 2000:
            zf.extract(name)
        
```

```python
#Делаем случайную выборку:

n = 22172457 - 1 #number of records in file
s = 221700 #desired sample size
filename = "goskatalog.csv"
skip = sorted(random.sample(range(1,n+1),n-s))
df = pd.read_csv(filename, skiprows=skip)
```

# Выгружаем только КОПУКи и делаем список уникальных значений

```python
gk_unique_kopuk = gk_kopuk['museum.code'].unique()
gk_unique_kopuk = pd.DataFrame(gk_unique_kopuk)
gk_unique_kopuk = gk_unique_kopuk.rename(columns={0: "museum.code"})
gk_unique_kopuk
```
